```python
import chess
target_white = "Rb5"
target_black = "Rb4"
def clean(san):
    return san.replace("x", "").replace("+", "").replace("#", "")
# standard point values; king is 0 since it's never captured
VALUES = {
    chess.PAWN: 1,
    chess.KNIGHT: 3,
    chess.BISHOP: 3,
    chess.ROOK: 5,
    chess.QUEEN: 9,
    chess.KING: 0,
}

def material(board, color):
    return sum(
        len(board.pieces(piece_type, color)) * value
        for piece_type, value in VALUES.items()
    )

threshold = 35
game_with_material = 0
game_with_arjun = 0
count = 0
with open('data/GM_filtered.pgn', encoding="utf-8") as infile, open('data/games_rb5_rb4.pgn', 'w', encoding="utf-8") as outfile:
    exporter = chess.pgn.FileExporter(outfile)
    while (game := chess.pgn.read_game(infile)) is not None:
        board = game.board()
        count += 1
        for move in game.mainline_moves():
            if board.turn == chess.WHITE and clean(board.san(move)) == target_white or board.turn == chess.BLACK and clean(board.san(move)) == target_black:
                total_material = material(board, chess.WHITE) + material(board, chess.BLACK)
                if total_material >= threshold:
                    game_with_material += 1
                    if game.headers["White"] == "Erigaisi Arjun" :
                        if board.turn == chess.WHITE and clean(board.san(move)) == target_white:
                            game_with_arjun += 1
                    elif game.headers["Black"] == "Erigaisi Arjun":
                        if board.turn == chess.BLACK and clean(board.san(move)) == target_black:
                            game_with_arjun += 1

            board.push(move)
        
        if count % 1000 == 0:
            print(count)
print(game_with_material)
print(game_with_arjun)
```
